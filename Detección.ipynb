# Análisis de histograma
En esta sección se compara el histograma de una imagen tomada a 200 MP vs:

 - La misma imagen pasada a Google Nano Banana Pro y generada por este modelo (edición de imagen)
 - La misma imagen pasada a ChatGPT 5.1 y generada por este modelo (edición de imagen)
 - Imagen generada a partir de un prompt por Google Nano Banana Pro (de texto a imagen)
 - Imagen generada a partir de un prompt por ChatGPT 5.1 (de texto a imagen)

```python
# Importación de librerías
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carga de todas las imágenes del dataset
img_real = cv2.imread('image.jpg')
img_fake1 = cv2.imread('Imágenes/test/futbol.jpeg')
img_fake2 = cv2.imread('Imágenes/test/realista_byn.png')
img_text1 = cv2.imread('Imágenes/test/chica_Realista.jpg')
img_text2 = cv2.imread('Imágenes/test/ext.jpeg')

# Conversión de BGR (formato OpenCV) a RGB para visualización correcta
imagen_rgb = cv2.cvtColor(img_real, cv2.COLOR_BGR2RGB)
imagen_rgb_fake = cv2.cvtColor(img_fake1, cv2.COLOR_BGR2RGB)
imagen_rgb_fake2 = cv2.cvtColor(img_fake2, cv2.COLOR_BGR2RGB)
image_text_rgb = cv2.cvtColor(img_text1, cv2.COLOR_BGR2RGB)
image_text_rgb2 = cv2.cvtColor(img_text2, cv2.COLOR_BGR2RGB)

print("Imágenes cargadas correctamente")
print(f"Dimensiones imagen real: {imagen_rgb.shape}")
```

Imagen original


```python
# Cálculo del histograma de la imagen original
histogram = cv2.calcHist([imagen_rgb], [0], None, [256], [0, 256])

# Visualización reducida para evitar problemas de memoria
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
ax.plot(histogram)
ax.set_title('Histograma - Imagen Original')
ax.set_xlabel('Valor de Píxel')
ax.set_ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

## 2. Conversión a Escala de Grises

Conversión de todas las imágenes RGB a escala de grises para normalizar el análisis de contraste y bordes.

```python
# Conversión a escala de grises usando transformación estándar RGB
gris_real = cv2.cvtColor(imagen_rgb, cv2.COLOR_RGB2GRAY)
gris_fake1 = cv2.cvtColor(imagen_rgb_fake, cv2.COLOR_RGB2GRAY)
gris_fake2 = cv2.cvtColor(imagen_rgb_fake2, cv2.COLOR_RGB2GRAY)
gris_text1 = cv2.cvtColor(image_text_rgb, cv2.COLOR_RGB2GRAY)
gris_text2 = cv2.cvtColor(image_text_rgb2, cv2.COLOR_RGB2GRAY)

# Visualización de las imágenes en escala de grises
fig, axes = plt.subplots(1, 5, figsize=(20, 4))
imagenes_gris = [gris_real, gris_fake1, gris_fake2, gris_text1, gris_text2]
titulos = ['Real', 'Nano Banana Pro Editada', 'GPT 5.1 Editada', 'Nano Banana Pro Texto', 'GPT 5.1 Texto']

for ax, img_gris, titulo in zip(axes, imagenes_gris, titulos):
    ax.imshow(img_gris, cmap='gray')
    ax.set_title(titulo)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## 3. Filtro Laplaciano - Detección de Bordes

Aplicación de máscara Laplaciana para realce de bordes y detección de alta frecuencia.

```python
# Definición del kernel Laplaciano 3x3
kernel_lap = np.array([[0, -1, 0],
                       [-1, 4, -1],
                       [0, -1, 0]])

# Aplicación del filtro Laplaciano a cada imagen en escala de grises
lap_real = cv2.filter2D(gris_real, -1, kernel_lap)
lap_fake1 = cv2.filter2D(gris_fake1, -1, kernel_lap)
lap_fake2 = cv2.filter2D(gris_fake2, -1, kernel_lap)
lap_text1 = cv2.filter2D(gris_text1, -1, kernel_lap)
lap_text2 = cv2.filter2D(gris_text2, -1, kernel_lap)

# Cálculo de métricas de alta frecuencia (promedio de magnitudes absolutas)
hf_real = np.mean(np.abs(lap_real))
hf_fake1 = np.mean(np.abs(lap_fake1))
hf_fake2 = np.mean(np.abs(lap_fake2))
hf_text1 = np.mean(np.abs(lap_text1))
hf_text2 = np.mean(np.abs(lap_text2))

print("Métricas de Alta Frecuencia:")
print(f"Real: {hf_real:.2f}")
print(f"Nano Banana Pro Editada: {hf_fake1:.2f}")
print(f"GPT 5.1 Editada: {hf_fake2:.2f}")
print(f"Nano Banana Pro Texto: {hf_text1:.2f}")
print(f"GPT 5.1 Texto: {hf_text2:.2f}")
```

```python
# Visualización de los mapas Laplacianos
fig, axes = plt.subplots(1, 5, figsize=(20, 4))
imagenes_lap = [lap_real, lap_fake1, lap_fake2, lap_text1, lap_text2]
titulos = ['Real', 'Nano Banana Pro Editada', 'GPT 5.1 Editada', 'Nano Banana Pro Texto', 'GPT 5.1 Texto']

for ax, img_lap, titulo in zip(axes, imagenes_lap, titulos):
    ax.imshow(img_lap, cmap='gray')
    ax.set_title(titulo)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## 4. Análisis Espectral - FFT 2D

Transformada de Fourier 2D para analizar la distribución de frecuencias y detectar patrones repetitivos característicos de síntesis IA.

```python
# Reducción de tamaño para análisis FFT (por limitación de memoria)
# Se mantiene la proporción de aspecto
scale_factor = 0.1
gris_real_small = cv2.resize(gris_real, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_AREA)
gris_fake1_small = cv2.resize(gris_fake1, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_AREA)
gris_fake2_small = cv2.resize(gris_fake2, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_AREA)
gris_text1_small = cv2.resize(gris_text1, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_AREA)
gris_text2_small = cv2.resize(gris_text2, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_AREA)

# Aplicación de FFT 2D a cada imagen reducida
f_real = np.fft.fft2(gris_real_small)
f_fake1 = np.fft.fft2(gris_fake1_small)
f_fake2 = np.fft.fft2(gris_fake2_small)
f_text1 = np.fft.fft2(gris_text1_small)
f_text2 = np.fft.fft2(gris_text2_small)

# Desplazar componente DC al centro del espectro
fshift_real = np.fft.fftshift(f_real)
fshift_fake1 = np.fft.fftshift(f_fake1)
fshift_fake2 = np.fft.fftshift(f_fake2)
fshift_text1 = np.fft.fftshift(f_text1)
fshift_text2 = np.fft.fftshift(f_text2)

# Calcular espectro de magnitud en escala logarítmica
mag_real = 20*np.log(np.abs(fshift_real)+1)
mag_fake1 = 20*np.log(np.abs(fshift_fake1)+1)
mag_fake2 = 20*np.log(np.abs(fshift_fake2)+1)
mag_text1 = 20*np.log(np.abs(fshift_text1)+1)
mag_text2 = 20*np.log(np.abs(fshift_text2)+1)

print("Espectros de frecuencia calculados correctamente")
```

```python
# Visualización de los espectros de frecuencia
fig, axes = plt.subplots(1, 5, figsize=(20, 4))
espectros = [mag_real, mag_fake1, mag_fake2, mag_text1, mag_text2]
titulos = ['Real', 'Nano Banana Pro Editada', 'GPT 5.1 Editada', 'Nano Banana Pro Texto', 'GPT 5.1 Texto']

for ax, espectro, titulo in zip(axes, espectros, titulos):
    ax.imshow(espectro, cmap='gray')
    ax.set_title(f'Espectro FFT - {titulo}')
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## 5. Cuantificación de Ruido Espectral

Análisis cuantitativo de la distribución de energía en el dominio de frecuencias. Se calculan métricas que caracterizan el nivel de ruido y uniformidad del espectro.

```python
# Cálculo de métricas espectrales para cada imagen
def calcular_metricas_espectrales(fshift):
    """
    Calcula métricas de ruido espectral:
    - Energía total del espectro
    - Energía en altas frecuencias (anillo exterior)
    - Ratio de energía alta/baja frecuencia
    """
    magnitud = np.abs(fshift)
    h, w = magnitud.shape
    cy, cx = h // 2, w // 2
    
    # Crear máscara para bajas frecuencias (círculo central, radio 15% del tamaño)
    y, x = np.ogrid[:h, :w]
    radio_bajo = min(h, w) * 0.15
    mascara_bajo = ((x - cx)**2 + (y - cy)**2) <= radio_bajo**2
    
    # Energía total
    energia_total = np.sum(magnitud**2)
    
    # Energía en bajas frecuencias
    energia_baja = np.sum(magnitud[mascara_bajo]**2)
    
    # Energía en altas frecuencias
    energia_alta = energia_total - energia_baja
    
    # Ratio alta/baja frecuencia
    ratio_af_bf = energia_alta / energia_baja if energia_baja > 0 else 0
    
    return {
        'energia_total': energia_total,
        'energia_alta': energia_alta,
        'energia_baja': energia_baja,
        'ratio_af_bf': ratio_af_bf
    }

# Calcular métricas para todas las imágenes
metricas_real = calcular_metricas_espectrales(fshift_real)
metricas_fake1 = calcular_metricas_espectrales(fshift_fake1)
metricas_fake2 = calcular_metricas_espectrales(fshift_fake2)
metricas_text1 = calcular_metricas_espectrales(fshift_text1)
metricas_text2 = calcular_metricas_espectrales(fshift_text2)

# Mostrar resultados
print("Métricas Espectrales - Ratio Alta/Baja Frecuencia:")
print(f"Real:                    {metricas_real['ratio_af_bf']:.4f}")
print(f"Nano Banana Pro Editada: {metricas_fake1['ratio_af_bf']:.4f}")
print(f"GPT 5.1 Editada:         {metricas_fake2['ratio_af_bf']:.4f}")
print(f"Nano Banana Pro Texto:   {metricas_text1['ratio_af_bf']:.4f}")
print(f"GPT 5.1 Texto:           {metricas_text2['ratio_af_bf']:.4f}")
print("\nInterpretación: Valores más altos indican mayor contenido de alta frecuencia (ruido/detalle)")
```

```python
# Métricas espectrales avanzadas
def calcular_metricas_avanzadas(fshift):
    """
    Métricas adicionales para detectar patrones de síntesis por IA:
    - Uniformidad radial: mide simetría del espectro
    - Entropía espectral: regularidad de la distribución de energía
    - Energía en frecuencias medias: detecta patrones en banda específica
    """
    magnitud = np.abs(fshift)
    h, w = magnitud.shape
    cy, cx = h // 2, w // 2
    
    # 1. Uniformidad radial: variación de energía por anillos concéntricos
    max_radio = min(h, w) // 2
    num_anillos = 20
    energias_anillos = []
    
    for i in range(num_anillos):
        r_inner = (i / num_anillos) * max_radio
        r_outer = ((i + 1) / num_anillos) * max_radio
        
        y, x = np.ogrid[:h, :w]
        dist = np.sqrt((x - cx)**2 + (y - cy)**2)
        mascara_anillo = (dist >= r_inner) & (dist < r_outer)
        
        if np.sum(mascara_anillo) > 0:
            energia_anillo = np.mean(magnitud[mascara_anillo])
            energias_anillos.append(energia_anillo)
    
    uniformidad_radial = np.std(energias_anillos) / (np.mean(energias_anillos) + 1e-10)
    
    # 2. Entropía espectral normalizada
    magnitud_norm = magnitud / (np.sum(magnitud) + 1e-10)
    magnitud_norm = magnitud_norm[magnitud_norm > 0]
    entropia = -np.sum(magnitud_norm * np.log(magnitud_norm + 1e-10))
    
    # 3. Energía en frecuencias medias (banda 15%-40% del radio)
    y, x = np.ogrid[:h, :w]
    dist = np.sqrt((x - cx)**2 + (y - cy)**2)
    r_min = max_radio * 0.15
    r_max = max_radio * 0.40
    mascara_media = (dist >= r_min) & (dist < r_max)
    
    energia_media = np.sum(magnitud[mascara_media]**2)
    energia_total = np.sum(magnitud**2)
    ratio_media = energia_media / energia_total if energia_total > 0 else 0
    
    return {
        'uniformidad_radial': uniformidad_radial,
        'entropia': entropia,
        'ratio_freq_media': ratio_media
    }

# Calcular métricas avanzadas
adv_real = calcular_metricas_avanzadas(fshift_real)
adv_fake1 = calcular_metricas_avanzadas(fshift_fake1)
adv_fake2 = calcular_metricas_avanzadas(fshift_fake2)
adv_text1 = calcular_metricas_avanzadas(fshift_text1)
adv_text2 = calcular_metricas_avanzadas(fshift_text2)

print("\n=== MÉTRICAS ESPECTRALES AVANZADAS ===\n")

print("1. UNIFORMIDAD RADIAL (menor = más uniforme/sospechoso):")
print(f"   Real:                    {adv_real['uniformidad_radial']:.4f}")
print(f"   Nano Banana Pro Editada: {adv_fake1['uniformidad_radial']:.4f}")
print(f"   GPT 5.1 Editada:         {adv_fake2['uniformidad_radial']:.4f}")
print(f"   Nano Banana Pro Texto:   {adv_text1['uniformidad_radial']:.4f}")
print(f"   GPT 5.1 Texto:           {adv_text2['uniformidad_radial']:.4f}")

print("\n2. ENTROPÍA ESPECTRAL (mayor = más desorden/natural):")
print(f"   Real:                    {adv_real['entropia']:.4f}")
print(f"   Nano Banana Pro Editada: {adv_fake1['entropia']:.4f}")
print(f"   GPT 5.1 Editada:         {adv_fake2['entropia']:.4f}")
print(f"   Nano Banana Pro Texto:   {adv_text1['entropia']:.4f}")
print(f"   GPT 5.1 Texto:           {adv_text2['entropia']:.4f}")

print("\n3. RATIO FRECUENCIAS MEDIAS (banda 15%-40%):")
print(f"   Real:                    {adv_real['ratio_freq_media']:.4f}")
print(f"   Nano Banana Pro Editada: {adv_fake1['ratio_freq_media']:.4f}")
print(f"   GPT 5.1 Editada:         {adv_fake2['ratio_freq_media']:.4f}")
print(f"   Nano Banana Pro Texto:   {adv_text1['ratio_freq_media']:.4f}")
print(f"   GPT 5.1 Texto:           {adv_text2['ratio_freq_media']:.4f}")
```
